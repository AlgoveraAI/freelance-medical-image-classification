```python
path = Path('/content/drive/MyDrive/colab_notebooks/algovera/nCight')
```

```python
len(img_fns)
```

```python
fig, axes = plt.subplots(10, 2, figsize=(20,60))

for i, ax in enumerate(axes.flatten()):
    fn = img_fns[i]
    ax.imshow(PILImage.create(fn))
    ax.axis('off')
    ax.title.set_text(str(fn).split()[1].split('/')[1])
```

```python
shapes = []
for i, ax in enumerate(axes.flatten()):
    fn = img_fns[i]
    shapes.append(PILImage.create(fn).shape)
```

```python
shapes, len(shapes)
```

```python
def get_json(fn):
    if fn.suffix == '.png':
        return f'{str(fn).split(".png")[0]}.json'

    elif fn.suffix == '.jpeg':
        return f'{str(fn).split(".jpeg")[0]}.json'
```

```python
for fn in img_fns:
    with open(get_json(fn), 'r') as f:
        print(json.load(f))
```
